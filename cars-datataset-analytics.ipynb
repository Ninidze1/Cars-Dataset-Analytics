```python
# სასურველი სტრიქონების და სვეტების ამოღება
desired_rows = df.iloc[:100]
desired_columns = desired_rows[['Name', 'Mileage', 'Year', 'Price', "Kilometers_Driven", "Transmission", "Power", "Owner_Type", "Mileage", "Seats"]]

desired_columns
```

```python
# "Name" სვეტის ინდექსად დაყენება
indexed_df = desired_columns.set_index('Name')
indexed_df
```

```python
# გაფილტვრა ორი პარამეტრით
filtered_df = indexed_df[(indexed_df['Kilometers_Driven'] < 40000) & (indexed_df['Year'] >= 2017)]

filtered_df
```

```python
# ორი პარამეტრით გაფილტრვა
sorted_df = filtered_df.sort_values(by=['Year', 'Price'], ascending=[False, True])

sorted_df
```

```python
# "Price" სვეტის სტატისტიკური ფუნქციების გაანგარიშება
mean_price = df['Price'].mean()
std_dev_price = df['Price'].std()
median_price = df['Price'].median()
min_price = df['Price'].min()
max_price = df['Price'].max()

mean_price, std_dev_price, median_price, min_price, max_price
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# CSV ფაილის ჩატვირთვა
df = pd.read_csv("train-data.csv")

# ფიგურისა და ღერძის განსაზღვრა
fig, ax = plt.subplots(2, 1, figsize=(12, 12))

# Bar chart საშუალო ფასისთვის Fuel_Type მიხედვით
fuel_types = df['Fuel_Type'].unique()
average_prices = [df[df['Fuel_Type'] == fuel]['Price'].mean() for fuel in fuel_types]

ax[0].bar(fuel_types, average_prices, color=['red', 'blue', 'green', 'purple', 'orange'])
ax[0].set_title('მანქანების საშუალო ფასი საწვავის ტიპის მიხედვით')
ax[0].set_ylabel('საშუალო ფასი')
ax[0].set_xlabel('საწვავის ტიპი')

# Line chart საშუალო ფასების ტენდენცია წლების განმავლობაში
years = np.sort(df['Year'].unique())
yearly_avg_prices = [df[df['Year'] == year]['Price'].mean() for year in years]

ax[1].plot(years, yearly_avg_prices, '-o', color='blue')
ax[1].set_title('მანქანების საშუალო ფასების ტენდენცია წლების განმავლობაში')
ax[1].set_ylabel('საშუალო ფასი')
ax[1].set_xlabel('წელი')

plt.tight_layout()
plt.show()
```
